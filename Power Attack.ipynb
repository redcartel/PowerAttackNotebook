```python
import numpy as np
import pandas as pd
```

```python
def hit(dc):
    roll = np.random.randint(1,21)
    if roll >= 20:
        return True
    if roll >= dc:
        return True
    return False

def advantage_hit(dc):
    roll = max(np.random.randint(1,21,2))
    if roll >= 20:
        return True
    if roll >= dc:
        return True
    return False
```

```python
def differential(dc, damage):
    _hit = hit(dc)
    p_hit = hit(dc+5)
    
    if _hit:
        d = damage
    else:
        d = 0
    
    if p_hit:
        pd = damage + 10
    else:
        pd = 0
    
    return pd - d

def adv_differential(dc, damage):
    _hit = advantage_hit(dc)
    p_hit = advantage_hit(dc + 5)
    
    if _hit:
        d = damage
    else:
        d = 0
    
    if p_hit:
        pd = damage+10
    else:
        pd = 0
    
    return pd - d
```

```python
def monty_diff(dc, damage, ntimes):
    return sum(differential(dc, damage) for _ in range(ntimes)) / ntimes

def monty_a_diff(dc, damage, ntimes):
    return sum(adv_differential(dc, damage) for _ in range(ntimes)) / ntimes
```

```python
d_a = [[monty_diff(dc, dam+.5, 20000) for dc in range(5,21)] for dam in range(5, 12)]
```

```python
d_aa = [[monty_a_diff(dc, dam+.5, 20000) for dc in range(5,21)] for dam in range(5, 12)]
```

```python
df = pd.DataFrame(d_a, index=[dam+.5 for dam in range(5, 12)], columns=range(5,21))
```

```python
dfa = pd.DataFrame(d_aa, index=[dam+.5 for dam in range(5,12)], columns=range(5,21))
```
